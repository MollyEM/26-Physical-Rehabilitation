```python
import csv
import os,random
import numpy as np

#given the folder of movements, take all of the files, and put into an array
def combine(dir_path, name):
    frames = 40
    filenames = []
    #get all of the names for the files inside the folder
    for path in os.listdir(dir_path):
        # check if current path is a file
        if os.path.isfile(os.path.join(dir_path, path)):
            filenames.append(path)
            
    
    length = len(filenames)
    
    #array to put all movements in folder into
    array = np.zeros(shape = (length * 75, 40))
    e = 0
    for p in filenames:
        print(p)
        #array for what to read in
        oneEp = np.zeros(shape = (40, 50))
        #open file (40 frames (rows) with 50 joint positions(cols))
        #append each file to its own list
        f = open(dir_path + '/' + p, 'r')
        c = csv.reader(f, delimiter = ',', quoting = csv.QUOTE_NONE)
        i = 0
        for row in c:
            oneEp[i, :] = row
            i += 1
        
        
        j = range(2, 51, 2)
        k = 0 #[0] * frames
        oneEp = np.insert(oneEp, j, k, axis = 1)
        oneEp_2 = np.swapaxes(oneEp, 1, 0)

        #add whole list to list of arrays of all episodes in the folder
        array[75 * e:75 * (e + 1), :] = oneEp_2
        e += 1
        f.close()
    
    #array = np.swapaxes(array, 1, 0)
    Data_mean = np.tile(np.mean(array, axis=1, keepdims=True), (1, array.shape[1]))

    # Compute centered Correct_Xm data
    array = array - Data_mean
    
    scaling_value = np.ceil(np.max(np.abs([np.max(array), np.min(array)])))
    array = array / scaling_value
    
    #array = np.swapaxes(array, 1, 0)

    #write to a csv
    f = open("C:/Users/Molly Meadows/OneDrive - University of Idaho/Documents/2023-2024/Capstone Project/26-Physical-Rehabilitation/Retest of videos/same_size_csvs" + '/InlineLungeReshaped_' + name + '.csv', 'w', newline = '')
    w = csv.writer(f)
    np.apply_along_axis(w.writerow, axis = 1, arr = array)
    f.close()
    
    
```
