```python
import cv2

# Load the OpenPose model
net = cv2.dnn.readNetFromCaffe('pose_deploy.prototxt', 'pose_iter_584000.caffemodel')

img = cv2.imread('Gabe.jpg')
frameWidth = img.shape[1]
frameHeight = img.shape[0]


inHeight = 368
inWidth = int((inHeight/frameHeight)*frameWidth)
blob = cv2.dnn.blobFromImage(img, 1.0/1000 , (inWidth, inHeight),
                          (0, 0, 0), swapRB=False, crop=False)

print(frameWidth, frameHeight)
print(inHeight/frameHeight, (inHeight/frameHeight)*frameWidth, 368*inHeight/frameHeight, 386*(inHeight/frameHeight)*frameWidth)

# Set the input for the network
net.setInput(blob)

# Run forward pass to get the pose estimation
output = net.forward()



# Define the keypoint mapping for OpenPose
keypoints_mapping = {
    0:  "Nose", 1:  "Neck", 2:  "RShoulder", 3:  "RElbow", 4:  "RWrist", 5:  "LShoulder", 6:  "LElbow",
    7:  "LWrist", 8:  "MidHip", 9:  "RHip", 10: "RKnee", 11: "RAnkle", 12: "LHip", 13: "LKnee",
    14: "LAnkle", 15: "REye", 16: "LEye", 17: "REar", 18: "LEar", 19: "LBigToe", 20: "LSmallToe",
    21: "LHeel", 22: "RBigToe", 23: "RSmallToe", 24: "RHeel", 25: "Background"
}
print("Checkpoint")
print(keypoints_mapping)
# Extract joint locations
joint_locations = []
for i in range(len(keypoints_mapping)):
    keypoint = output[0, i, :, :]
    min_val, confidence, min_loc, point = cv2.minMaxLoc(keypoint)
    print(min_val, confidence, min_loc, point)
    if confidence > 0.3:  # You can adjust the confidence threshold as needed
        joint_locations.append((9*int(point[0]), 9*int(point[1])))
        print("append ", i," , ", point)
    else:
        joint_locations.append(None)
        
print(len(joint_locations))

# Now, joint_locations contains the locations of the detected joints

print("Checkpoint2")
for location in joint_locations:
    if location:
        cv2.circle(img, location, 5, (0, 0, 255), -1)

# Display the image with joint locations
cv2.imshow('Pose Estimation', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
